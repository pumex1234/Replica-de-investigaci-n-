## 1. Visualización de Modelos Predichos para T1226-D1
En el paper, T1226-D1 es un objetivo particularmente desafiante. La predicción estándar falla produciendo una hélice alfa artificialmente extendida, mientras que MULTICOM (gracias al uso extensivo de AF3) logra una forma compacta correcta.
A continuación visualizamos un modelo ColabFold (Server config) versus la mejor predicción (AlphaFold3).

```python
def view_pdb(path, color='spectrum'):
    if not os.path.exists(path):
        print(f'Error: archivo no encontrado {path}')
        return
    with open(path, 'r') as f:
        pdb_data = f.read()
    view = py3Dmol.view(width=400, height=300)
    view.addModel(pdb_data, 'pdb')
    view.setStyle({'cartoon': {'color': color}})
    view.zoomTo()
    view.show()

# Modelo del servidor estándar que suele fallar (extendido)
failed_model = r'zenodo/Predictions/T1226/colabfold_web_0.pdb'
print("Modelo ColabFold/Base (predicción extendida errónea):")
view_pdb(failed_model, color='red')

# Modelo AlphaFold3 exitoso de MULTICOM (compacto)
success_model = r'zenodo/Predictions/T1226/af3_pawan_0.pdb'
print("Modelo AlphaFold3 Interno (predicción compacta correcta):")
view_pdb(success_model, color='blue')
```
